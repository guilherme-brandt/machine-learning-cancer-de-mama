```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#lê dataset
df = pd.read_csv('data.csv')
```

```python
#conta os valores de diagnosis
df['diagnosis'].value_counts()
```

```python
#conta os valores nulos
df.isnull().sum()
```

```python
#remove valores nulos
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
#variáveis indepententes
x = df.drop('diagnosis',axis=1)
#variáveis dependentes (alvo)
y = df.diagnosis
```

```python
from sklearn.preprocessing import LabelEncoder
#cria o objeto
lb = LabelEncoder()
y = lb.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
#separa conjuntos de teste e treino
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
#padroniza as escalas
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
#inicializa modelo de rede neural
classifier = Sequential()
```

```python
#primeira camada escondida da rede neural
classifier.add(Dense(units=9,kernel_initializer='he_uniform',activation='relu',input_dim=30))
#segunda
classifier.add(Dense(units=9,kernel_initializer='he_uniform',activation='relu'))
#terceira
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
#resumo das camadas
classifier.summary()
```

```python
#monta a rede neural
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#ensina o modelo com o conjunto de treino
model = classifier.fit(xtrain,ytrain,batch_size=100,epochs=100)
```

```python
#testa a rede neural com o conjunto de teste
y_pred = classifier.predict(xtest)
```

```python
#converte os valores numéricos para true e false
y_pred = (y_pred>0.5)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
#matriz de confusão e escore para comparar resultados reais(ytest) dos obtidos pelo modelo(y_pred)
cm = confusion_matrix(ytest,y_pred)
score = accuracy_score(ytest,y_pred)
print(cm)
print('score is:',score)
```
